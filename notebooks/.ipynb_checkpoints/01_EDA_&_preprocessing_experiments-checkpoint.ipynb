## Preprocessing

### 1- Remove Duplicates

```python
df = df.drop_duplicates()
```

### 2- Encode categorical variables

```python
le = LabelEncoder()
df['gender'] = le.fit_transform(df['gender'])
df['smoking_history'] = le.fit_transform(df['smoking_history'])
```

### 3-Handle outliers

```python
def cap_outliers(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    data[column] = data[column].clip(lower_bound, upper_bound)
    return data

for col in ['age', 'bmi', 'HbA1c_level', 'blood_glucose_level']:
    data = cap_outliers(df, col)
```

### 4- Scale numerical features

```python
scaler = StandardScaler()
data[['age', 'bmi', 'HbA1c_level', 'blood_glucose_level']] = scaler.fit_transform(
    data[['age', 'bmi', 'HbA1c_level', 'blood_glucose_level']]
)
```

```python
data.dtypes
```

```python
print("Generating Pair Plot...")
sns.pairplot(data[['age', 'bmi', 'HbA1c_level', 'blood_glucose_level', 'diabetes']], 
             hue='diabetes', diag_kind='kde')
plt.suptitle('Pair Plot of Numerical Features by Diabetes', y=1.02)
plt.show()
```

```python
print("Generating Box Plots...")
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
numerical_cols = ['age', 'bmi', 'HbA1c_level', 'blood_glucose_level']
for i, col in enumerate(numerical_cols):
    sns.boxplot(x='diabetes', y=col, data=data, ax=axes[i//2, i%2])
    axes[i//2, i%2].set_title(f'{col} vs. Diabetes')
plt.tight_layout()
plt.show()
```

```python
print("Generating Count Plots...")
categorical_cols = ['gender', 'smoking_history', 'hypertension', 'heart_disease']
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
for i, col in enumerate(categorical_cols):
    sns.countplot(x=col, hue='diabetes', data=data, ax=axes[i//2, i%2])
    axes[i//2, i%2].set_title(f'{col} vs. Diabetes')
plt.tight_layout()
plt.show()
```
